```python
from os.path import isdir, join
from os import mkdir, listdir
import numpy as np
from PIL import Image

if not isdir('train'):
    mkdir('train')
if not isdir('train/patched_images'):
    mkdir('train/patched_images')
if not isdir('train/patched_masks'):
    mkdir('train/patched_masks')
if not isdir('train/patched_images/images'):
    mkdir('train/patched_images/images')
if not isdir('train/patched_masks/masks'):
    mkdir('train/patched_masks/masks')


    

img_root = 'MoNuSeg 2018 Training Data/Tissue Images'
mask_root = 'MoNuSeg 2018 Training Data/Binary_masks'

patched_img__dir = 'train/patched_images/images'
patched_masks_dir = 'train/patched_masks/masks'

img_list = [f[: -4] for f in listdir(img_root)]

for image in tqdm(img_list, desc = 'Creating train patches ...'):
    im = np.array(Image.open(join(img_root, image + '.tif')).resize((1024, 1024), Image.NEAREST), dtype = np.uint8)
    ms = np.array(Image.open(join(mask_root, image + '.png')).resize((1024, 1024), Image.NEAREST), dtype = np.uint8)
    

    count = 0
    for i in range(0, 769, 256):
        for j in range(0, 769, 256):
            img_patch = Image.fromarray(im[0 + i: 256 + i, 0 + j: 256 + j])
            msk_patch = Image.fromarray(ms[0 + i: 256 + i, 0 + j: 256 + j])
            img_patch.save(patched_img__dir + '/' + image + '_' + str(count) + '.png')
            msk_patch.save(patched_masks_dir + '/' + image + '_' + str(count) + '.png')
            
            count += 1
            
print('\n Train patches successfully generated!')
```
